```python
import numpy as np    #importing numpy and pandas packages
import pandas as pd
df = pd.read_csv("data.csv")        #reading csv file
```

```python
df.drop(columns='Unnamed: 0',axis=1,inplace=True)       #Deleting unnamed columns
```

```python
#showing distribution of male & female
gender_categ=pd.Categorical(df['Gender']) #using categorical attribute to categorize them
gender_categ.describe() #describing the categorical
```

```python
df['Name'].value_counts().head()  #Showing the top 5 most preferred names
```

```python
#Showing the median name occurence in the dataset
t=df.median().Id   #getting Id of the median record 
name = df[df['Id']==t].Name  #getting name of the median record
Na=list(name)[0]  
df1 = df[df['Name']==Na]   #finding dataframe with name is equal to 'Na'
df1['Name'].value_counts() #showing count of the name
```

```python
#Showing distribution of male and female born count by states
df1=df.groupby(['State','Gender']) #use groupby to group based on state and gender
df1.count() #printing count using grouped based
```
